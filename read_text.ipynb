```python
import cv2
import easyocr
import matplotlib.pyplot as plt
import numpy as np
```

##Rasmni o'qish

```python
!wget https://raw.githubusercontent.com/Davron030901/Passport_and_ID-karta_read/main/text-detection-python-tesseract-easyocr-textract/uzb_passport.jpg
```

```python
image_path = 'uzb_passport.jpg'

img = cv2.imread(image_path)
```

```python
reader = easyocr.Reader(['uz'], gpu=True)

text_ = reader.readtext(img)
```

```python
threshold = 0.25
for t_, t in enumerate(text_):
    print(t)

    bbox, text, score = t

    if score > threshold:
        cv2.putText(img, text, [int(j) for j in bbox[0]], cv2.FONT_HERSHEY_COMPLEX, 2.5, (0, 0, 255), 5)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image_path = 'ID_karta.jpg'
img = cv2.imread(image_path)
```

```python
reader = easyocr.Reader(['en'], gpu=True)
text_ = reader.readtext(img)
```

```python
threshold = 0.25
word_dict = {}
all_words = []

for t_, t in enumerate(text_):
    bbox, text, score = t
    if score > threshold:
        words = text.split()
        all_words.extend(words)
```

## Lug'atni yaratish


```python
for i in range(len(all_words) - 1):
    word_dict[all_words[i]] = all_words[i+1]
```

##Natijani chop etish

```python
print("So'zlar lug'ati:")
for key, value in word_dict.items():
    print(f"'{key}': '{value}'")
```
