# OCR ni ishga tushirish


```python
def process_image(image_path, lang='uz', gpu=False):
    img = cv2.imread(image_path)
    reader = easyocr.Reader([lang], gpu=gpu)
    text_ = reader.readtext(img)
    return text_, img
```

# Ma'lumotlarni ajratish

```python
def extract_info(text, pattern):
    match = re.search(pattern, text)
    return match.group() if match else None
```

# Passport uchun rasmni yuklash


```python
download_image(
    "https://raw.githubusercontent.com/Davron030901/Passport_and_ID-karta_read/main/text-detection-python-tesseract-easyocr-textract/uzb_passport.jpg",
    "uzb_passport.jpg"
)
```

# OCR ni ishga tushirish

```python
text_, img = process_image("uzb_passport.jpg", lang='uz', gpu=torch.cuda.is_available())
```

# Natijalarni chop etish

```python
threshold = 0.25
for bbox, text, score in text_:
    if score > threshold:
        print(f"Aniqlangan matn: {text}")
        cv2.putText(img, text, [int(j) for j in bbox[0]], cv2.FONT_HERSHEY_COMPLEX, 2.5, (0, 0, 255), 5)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
